```python
# Average Salary by Job Title:
avg_salary_by_title = jobs_df.groupby('job_title')['salary_in_usd'].mean().sort_values(ascending=False)
print(avg_salary_by_title)
```

```python
# Average classification of job role:
avg_salary_by_job_category = jobs_df.groupby(['work_year','job_category'])['salary_in_usd'].mean()
print(avg_salary_by_job_category)
```

```python
# what company location are represented?
jobs_df['company_location'].value_counts()
```

```python
# Average job salary by company location:
avg_salary_by_company_location = jobs_df.groupby('company_location')['salary_in_usd'].mean().sort_values(ascending=False)
print(avg_salary_by_company_location)
```

```python
# whats the type of compnay environment are represented?
jobs_df['work_settings'].value_counts()
```

```python
# Average Salary by work environment:
avg_salary_by_work_setting = jobs_df.groupby(['work_year','work_setting'])['salary_in_usd'].mean()
print(avg_salary_by_work_setting)
```

```python
# Average Salary by experience level:
avg_salary_by_experience_level = jobs_df.groupby(['work_year','experience_level'])['salary_in_usd'].mean()
# sort_values(ascending=False)
print(avg_salary_by_experience_level)
```

## Visualize

Create any visualizations for your EDA here. Make note in the form of code comments of what your thought process is for your visualizations.

```python

# bar chart by job_category 
jobs_df['job_category'].value_counts().plot(kind='bar');

```

```python
# Multi Line chart - Average Salary by Work Environment
import numpy
years = numpy.sort(jobs_df["work_year"].unique())
print(years)
# avg_salary = jobs_df.groupby("work_year")["salary_in_usd"].mean()


hybrid_list = []
in_person_list = []
remote_list = []

# in billions
for y in years:
    hybrid = jobs_df.loc[(jobs_df["work_year"] == y) & (jobs_df["work_setting"] == "Hybrid")]["salary_in_usd"].mean()
    in_person = jobs_df.loc[(jobs_df["work_year"] == y) & (jobs_df["work_setting"] == "In-person")]["salary_in_usd"].mean()
    remote = jobs_df.loc[(jobs_df["work_year"] == y) & (jobs_df["work_setting"] == "Remote")]["salary_in_usd"].mean()
    hybrid_list.append(hybrid)
    in_person_list.append(in_person)
    remote_list.append(remote)


plt.plot(years, hybrid_list, label = "Hybrid")
plt.plot(years, in_person_list, label = "In-person")
plt.plot(years, remote_list, label = "Remote")


plt.title("Average Salary by Work Environment")
plt.xlabel("Work Year")
plt.ylabel("Average Salary in USD")
plt.legend()

```

```python
# pie chart

mean_salary = jobs_df.groupby('work_setting')['salary_in_usd'].sum()
plt.pie(mean_salary, labels=mean_salary.index, autopct='%1.1f%%')
plt.title('Percentage of work environment')
plt.show()
```

```python
# pie chart

mean_salary = jobs_df.groupby('experience_level')['salary_in_usd'].mean()
plt.pie(mean_salary, labels=mean_salary.index, autopct='%1.1f%%')
plt.title('Percentage of salary by Experience level work environment')
plt.show()
```
